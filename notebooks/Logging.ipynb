```python
import logging
```

```python
name = 'John'
logger = logging.getLogger(__name__)

logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', datefmt='%d-%b-%Y %H:%M:%S', level=logging.DEBUG)  # Can only be called once
# logging.basicConfig(filename='app.log', filemode='w', format='%(name)s - %(levelname)s - %(message)s', datefmt='%c')
```

```python
logging.debug('This is an debug message')
logging.info('{}, this info will get logged'.format(name))
# logging.warning('This is a warning message')
# logging.error('This is an error message')
# logging.critical('This is a critical message')
```

```python
a = 5
b = 0

try:
    c = a / b
except ZeroDivisionError as z:
    # logging.error("Exception occurred", exc_info=True)
    logging.exception("Exception occurred")  # Only shows ERROR level
```

```python
a = '1'
b = 1

try:
    c = a + b
except TypeError as t:
    logging.exception("Cannot add incompatible types")  # Only shows ERROR level
```
