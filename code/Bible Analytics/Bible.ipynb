# Bible Analytics

Source: https://api.nlt.to/

```python
import json
import requests
import pandas as pd
import seaborn as sns
import spacy
from spacy import displacy
from getpass import getpass
from bs4 import BeautifulSoup as soup
```

```python
key = "c20cf5ba-88f1-4ad4-b902-f379b6afe3d0"
url = "https://api.nlt.to/api/passages?ref={}&key={}".format("1 Corinthians.1", key)
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.headers.get("content-type")
```

```python
data = response.text
```

```python
soup = soup(data, features="html.parser")
```

```python
soup
```

```python
text = soup.get_text()
text
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
displacy.serve(doc, style="ent")
```
